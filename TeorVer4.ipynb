```python
# Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
# Найдите ее среднее значение и дисперсию.

print('Задача 1')
print(f'Cреднее значение = {(200+800)/2}')
print(f'Дисперсия = {pow(800-200,2)/12}')

# О случайной непрерывной равномерно распределенной величине 
# B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу 
# величины B и ее среднее значение зная, что левая граница равна 0.5? 
# Если да, найдите ее.

right = pow(12*0.2,0.5)+0.5

print('Задача 2')
print(f'Правая граница = {round(right,5)}')
print(f'Дисперсия = {round(pow(right-0.5,2)/12,5)}')
print(f'Среднее значение = {round((right+0.5)/2,5)}')

# Непрерывная случайная величина X распределена нормально и задана 
# плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)  
# Найдите: а). M(X) б). D(X) в). std(X) (среднее квадратичное отклонение)

std_X = 4 # Из функции плотности распределения
M_X = -2 # Из функции плотности распределения
D_X = std_X*std_X

print('Задача 3')
print(f'а) M(X) = {M_X}')
print(f'б) D(X) = {D_X}')
print(f'в) std(X) (среднее квадратичное отклонение) = {std_X}')

# Рост взрослого населения города X имеет нормальное распределение. 
# Причем, средний рост равен 174 см, а среднее квадратичное отклонение 
# равно 8 см. Какова вероятность того, что случайным образом выбранный 
# взрослый человек имеет рост: а). больше 182 см б). больше 190 см 
# в). от 166 см до 190 см г). от 166 см до 182 см д). от 158 см до 190 см 
# е). не выше 150 см или не ниже 190 см 
# ж). не выше 150 см или не ниже 198 см з). ниже 166 см.

# z-показатель = (x - μ) / σ 
# μ = 174
# σ = 8 
#  Значения Z берутся из таблицы
print('Задача 4')
# а). больше 182 см
# z-показатель = (182 - 174) /8 = 1
# Z=0,84134
print(f'а). Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см = {round(1 - 0.84134,5)}')
# б). больше 190 см
# z-показатель = (190 - 174) /8 = 2
# Z=0,97725
print(f'б). Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см = {round(1 - 0.97725,5)}')
# в). от 166 см до 190 см
# z-показатель = (166 - 174) /8 = -1
# z-показатель = (190 - 174) /8 = 2
# Z=0,15866
# Z=0,97725
print(f'в). Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см = {round(0.97725-0.15866,5)}')
# г). от 166 см до 182 см
# z-показатель = (166 - 174) /8 = -1
# z-показатель = (182 - 174) /8 = 1
# Z=0,15866
# Z=0,84134
print(f'г). Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см = {round(0.84134-0.15866,5)}')
# д). от 158 см до 190 см 
# z-показатель = (158 - 174) /8 = -2
# z-показатель = (190 - 174) /8 = 2
# Z=0,02275
# Z=0,97725
print(f'д). Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см  = {round(0.97725-0.02275,5)}')
# е). не выше 150 см или не ниже 190 см 
# z-показатель = (150 - 174) /8 = -3
# z-показатель = (190 - 174) /8 = 2
# Z=0,00135
# Z=0,97725
print(f'е). Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см  = {round(1-0.97725+0.00135,5)}')
# ж). не выше 150 см или не ниже 198 см 
# z-показатель = (150 - 174) /8 = -3
# z-показатель = (198 - 174) /8 = 3
# Z=0,00135
# Z=0,99865
print(f'ж). Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см  = {round(1-0.99865+0.00135,5)}')
# з). ниже 166 см.
# z-показатель = (166 - 174) /8 = -1
# Z=0,15866
print(f'ж). Вероятность того, что случайным образом выбранный взрослый человек имеет рост  ниже 166 см = {0.15866}')

# На сколько сигм (средних квадратичных отклонений) отклоняется рост 
# человека, равный 190 см, от математического ожидания роста в 
# популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
# = (x - μ) / σ 
print('Задача 5')
print(f'На {(190-178)/pow(25,0.5)} σ (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см')
```
