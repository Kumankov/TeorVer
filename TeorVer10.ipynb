Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.
Даны значения роста в трех группах случайно выбранных спортсменов:

Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.

Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.

Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

*Для проведения post hoc теста выборки должны быть одинакового размера, поэтому дополним выборки футболистов и хоккеистов средними значениями в этих выборках, чтобы увеличение объема выборки не отражалось на среднем значении.*

*Футболисты1: 173, 175, 180, 178, 177, 185, 183, 182, 178.667, 178.667, 178.667.*

*Хоккеисты1: 177, 179, 180, 188, 177, 172, 171, 184, 180, 179.125, 179.125.*

*Штангисты1: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.*


*k = 3 (3 группы)*

*n = 33 (3 * 11)*

*Примем α =  0.05* 

*Cтепени свободы для определения табличного значения критерия Фишера 2 (k - 1) и 30 (n - k).*

*Табличный критерий Фишера между  3.30 и 3.34.*


```python
import numpy as np
import scipy.stats as stats

football = np.array([173, 175, 180, 178, 177, 185, 183, 182, 178.667, 178.667, 178.667])
hoсkey = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180, 179.125, 179.125])
barbell = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])

f = stats.f_oneway(football,hoсkey,barbell) 
print(f)
```

**Расчетный критерий Фишера равен 7.20353, что больше, чем табличный, который находится в диапазоне (3.34, 3.40). Следовательно, можно сделать вывод, что есть статистически значимые различия  среднего роста среди взрослых футболистов, хоккеистов и штангистов.**

**Для определения конкретных пар групп, между которыми наблюдаются различия, воспользуемся post hoc тестом Тьюки.**

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
df = pd.DataFrame({'height':[173, 175, 180, 178, 177, 185, 183, 182, 178.667, 178.667, 178.667,
                             177, 179, 180, 188, 177, 172, 171, 184, 180, 179.125, 179.125,
                             172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170],
                   'group': np.repeat(['football','hoсkey','barbell'], repeats = 11)})
tukey = pairwise_tukeyhsd(endog=df['height'],
                          groups=df['group'],
                          alpha=0.05)
print(tukey)
```

**Результаты выполнения теста Тьюки говорят, что наблюдаются различия между штангистами и футболистами, а также между штангистами и хоккеистами. Статистически значимых различий среднего роста между хоккеистами и футболистами не обнаружено.**

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
df = pd.DataFrame({'height':[173, 175, 180, 178, 177, 185, 183, 182,
                             177, 179, 180, 188, 177, 172, 171, 184, 180,
                             172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170],
                   'group': np.repeat(['football','hoсkey','barbell'], repeats = (8,9,11))})
tukey = pairwise_tukeyhsd(endog=df['height'],
                          groups=df['group'],
                          alpha=0.05)
print(tukey)
```
