```python
# Вероятность того, что стрелок попадет в мишень, 
# выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. 
# Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.
def fact(n):
  factorial = 1
  while n > 1:
    factorial *= n
    n -= 1 
  return(factorial)
# Решение по формуле Бернулли

prob1 = fact(100)/(fact(85)*fact(15))*pow(0.8,85)*pow(0.2,15)
print(f'Вероятность, что стрелок попадет в цель ровно 85 раз = {round(prob1,5)}')

# Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, 
# равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. 
# Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, 
# что перегорят ровно две?

# Решение по распределению Пуассона
lamb = 0.0004 * 5000
e_const = 2.71828

# А - ноль лампочек
prob2_A = pow(lamb,0)/fact(0)*pow(e_const,-lamb)
print(f'Вероятность, что ни одна лампочка не перегорит в первый день = {round(prob2_A,5)}')

# B - две лампочки
prob2_B = pow(lamb,2)/fact(2)*pow(e_const,-lamb)
print(f'Вероятность, что перегорят ровно 2 лампочки в первый день = {round(prob2_B,5)}')

# Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?
# Решение по формуле Бернулли
prob3 = fact(144)/(fact(70)*fact(74))*pow(0.5,70)*pow(0.5,74)
print(f'Вероятность, что орел выпадет ровно 70 раз = {round(prob3,5)}')

# В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, 
# из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. 
# Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? 
# Какова вероятность того, что хотя бы один мяч белый?

# A все мячи белые 
prob4_A = 7/10*6/9*9/11*8/10
print(f'Вероятность, что все мячи белые = {round(prob4_A,5)}')

# B ровно два мяча белые
# 6 вариантов: 2 белых из первого и не белее из второго, 2 не белых из первого и 2 белых из второго
# 4 варианта по одному белому и не белому из каждого ящика, в засисимости от того какой сначала вынимается из ящика
prob4_B = (7/10*6/9*2/11*1/10)+(3/10*2/9*9/11*8/10)+(7/10*3/9*9/11*2/10)+(7/10*3/9*2/11*9/10)+(3/10*7/9*9/11*2/10)+(3/10*7/9*2/11*9/10)
print(f'Вероятность, что ровно два мяча белые = {round(prob4_B,5)}')

# C хотя бы один мяч белый
# 1- вероятнось, что ни одного белого.
prob4_C = 1-3/10*2/9*2/11*1/10
print(f'Вероятность, что ни одного белого = {round(prob4_C,5)}')


```
