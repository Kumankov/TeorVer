```python
# Из колоды в 52 карты извлекаются случайным образом 4 карты. 
# a) Найти вероятность того, что все карты – крести. 
# б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.
def fact(n):
  factorial = 1
  while n > 1:
    factorial *= n
    n -= 1 
  return(factorial)
combs = fact(52)/(fact(4)*fact(48)) # Всего комбинаций из 4 карт
clubsCombs = fact(13)/(fact(4)*fact(9)) # Всего комбинаций из 4 крестей

print(f'Вероятность, что все крести = {round(clubsCombs/combs,5)}')
print(f'Вероятность, что все крести (второй вариант)= {round(13/52*12/51*11/50*10/49,5)}') # извлекаем по одной крестовой карте 4 раза 

notAceCombs = fact(48)/(fact(4)*fact(44)) # Комбинаций без тузов
print(f'Вероятность, что среди 4-х карт окажется хотя бы один туз = {round(1-notAceCombs/combs,5)}')
# Для второго варианта решения необходимо определить вероятность выпадения следующих комбинаций:
# 1 туз + 3 не туза, 2 туза + 2 не туза, 3 туза + 1 не туз, 4 туза 
treeCombs = fact(48)/(fact(3)*fact(45))# 3 не туза
twoCombs = fact(48)/(fact(2)*fact(46))# 2 не туза
oneCombs = 48# 1 не туз
oneAceCombs = 4# 1 туз
twoAceCombs = fact(4)/(fact(2)*fact(2)) #2 туза
threeAceCombs = fact(4)/(fact(1)*fact(3))# 3 туза
fourAceCombs = 1 # 4 туза
oneAceProb = oneAceCombs * treeCombs / combs + twoAceCombs * twoCombs / combs + threeAceCombs * oneCombs / combs + fourAceCombs / combs
print(f'Вероятность, что среди 4-х карт окажется хотя бы один туз (второй вариант) = {round(oneAceProb,5)}')
# события notAceCombs, oneAceCombs * treeCombs, twoAceCombs * twoCombs, threeAceCombs * oneCombs и fourAceCombs составляют полную группу событий
print(notAceCombs + oneAceCombs * treeCombs + twoAceCombs * twoCombs + threeAceCombs * oneCombs + fourAceCombs == combs)

# На входной двери подъезда установлен кодовый замок, содержащий десять кнопок 
# с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. 
# Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

doorCombs = fact(10)/(fact(3)*fact(7)) # Комбинаций из трех неповторяющихся цифр
print(f'Вероятность, что человек, не знающий код, откроет дверь с первой попытки = {round(1/doorCombs,5)}')

# Вероятность угадывания кода из 3 цифр, если важен порядок цифр = 1/10*1/9*1/8. Если порядок цифр не важен, 
# то вероятность угадывания равна сумме вероятностей 3! равновероятных событий угадывания 3 цифр.  
print(f'Вероятность, что человек, не знающий код, откроет дверь с первой попытки (второй вариант) = {round((1/10*1/9*1/8)*fact(3),5)}') 

# В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом 
# извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?
detCombs = fact(15)/(fact(3)*fact(12)) # количество вариантов по 3 детали
colorDetCombs = fact(9)/(fact(3)*fact(6)) # варианты 3 окрашенных деталей
print(f'Вероятность, что все извлеченные детали окрашены = {round(colorDetCombs/detCombs,5)}')
print(f'Вероятность, что все извлеченные детали окрашены (второй вариант) = {round(9/15 * 8/14 * 7/13,5)}') # извлекаем по 1 крашенной детали 3 раза

# В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, 
# что 2 приобретенных билета окажутся выигрышными?
lotCombs = fact(100)/(fact(2)*fact(98)) # количество вариантов по 2 любых билета
print(f'Вероятность, что 2 приобретенных билета окажутся выигрышными = {round(1/lotCombs,8)}')
print(f'Вероятность, что 2 приобретенных билета окажутся выигрышными (второй вариант) = {round(2/100 * 1/99,8)}') # выбираем по 1 выигрышному билету 2 раза.
```
