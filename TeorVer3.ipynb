```python
def fact(n):
  factorial = 1
  while n > 1:
    factorial *= n
    n -= 1 
  return(factorial)

# Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 
# Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, 
# среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
salary = (100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150)
average = sum(salary)/len(salary)
print('Задача 1')
print(f'Cреднее арифметическое = {average}')

square_S = 0
for i in salary:
  square_S += pow((i-average),2)
unbiased_variance = square_S / (len(salary) -1)
square_S /= len(salary)  
variance = square_S 
S = pow(square_S,0.5)
print(f'Среднее квадратичное отклонение = {round(S,5)}')
print(f'Смещенная оценка дисперсии = {round(variance,5)}')
print(f'Несмещенная оценка дисперсии = {round(unbiased_variance,5)}')

# В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
# Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

# 3 мяча можно получить следующими способами: 2 белых из первого ящика и 1 из второго; 1 белый из первого и 2 из второго;
# 0 белых из первого и 3 из второго

first_all = fact(8)/fact(6)/fact(2) # всего вариантов достать 2 мяча из первого ящика
second_all = fact(12)/fact(8)/fact(4) # всего вариантов достать 4 мяча из второго ящика

zero_white_first = fact(3)/fact(2)/fact(1) # вариантов достать 0 белых мячей из первого ящика
one_white_first = fact(5)/fact(4)/fact(1)*fact(3)/fact(2)/fact(1) # вариантов достать 1 белый мяч из первого ящика
two_white_first = fact(5)/fact(2)/fact(3) # вариантов достать 2 белых мяча из первого ящика
one_white_second = fact(5)/fact(4)/fact(1)*fact(7)/fact(4)/fact(3) # вариантов достать 1 белый мяч из второго ящика
two_white_second = fact(5)/fact(2)/fact(3)*fact(7)/fact(5)/fact(2) # вариантов достать 2 белых мяча из второго ящика
three_white_second = fact(5)/fact(2)/fact(3)*fact(7)/fact(6)/fact(1) # вариантов достать 3 белых мяча из второго ящика

p_tree_white = zero_white_first/first_all*three_white_second/second_all + one_white_first/first_all*two_white_second/second_all + two_white_first/first_all*one_white_second/second_all
print('Задача 2')
print(f'Вероятность вытащить 3 белых шара = {round(p_tree_white,5)}')


# На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого 
# спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: 
# a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

# Решение по формуле Байеса
# P(Bi|A) = P(A|Bi)*P(Bi)/P(A)
# P(A) = 1/3*0,9 + 1/3*0,8 +1/3*0,6

P_A = 1/3*0.9 + 1/3*0.8 + 1/3*0.6
P_A_B1 = 1/3*0.9
P_A_B2 = 1/3*0.8
P_A_B3 = 1/3*0.6

P_B1_A = P_A_B1/P_A 
P_B2_A = P_A_B2/P_A 
P_B3_A = P_A_B3/P_A 

print('Задача 3')
print(f'Вероятность, что выстрел произведён первым спортсменом = {round(P_B1_A,5)}')
print(f'Вероятность, что выстрел произведён вторым спортсменом = {round(P_B2_A,5)}')
print(f'Вероятность, что выстрел произведён третьим спортсменом = {round(P_B3_A,5)}')

# В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, 
# сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента 
# факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, 
# что он учится: a). на факультете A б). на факультете B в). на факультете C?

# решение аналогично предыдущей задаче
# P(A) = 1/4*0,8 + 1/4*0,7 + 2/4*0,9

p_A = 1/4*0.8 + 1/4*0.7 + 1/2*0.9
p_A_B1 = 1/4*0.8
p_A_B2 = 1/4*0.7
p_A_B3 = 1/2*0.9

p_B1_A = p_A_B1/p_A 
p_B2_A = p_A_B2/p_A 
p_B3_A = p_A_B3/p_A 

print('Задача 4')
print(f'Вероятность, что студент учится на факультете А  = {round(p_B1_A,5)}')
print(f'Вероятность, что студент учится на факультете B = {round(p_B2_A,5)}')
print(f'Вероятность, что студент учится на факультете C = {round(p_B3_A,5)}')

# Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, 
# для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: 
# а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

p_first_detail = 0.1
p_second_detail = 0.2
p_third_detail = 0.25
p_all = p_first_detail * p_second_detail * p_third_detail
# Вероятность того, что две детали выйдут из строя = сумме вероятностей того, что любые две выходят из стороя, а оставшаяся нет.
p_two = p_first_detail * p_second_detail * (1-p_third_detail) + p_first_detail * (1-p_second_detail) * p_third_detail + (1-p_first_detail) * p_second_detail * p_third_detail
#  Вероятность, что хотя бы одна деталь выйдет из строя расчитывается как 1 минус вероятность, что ни одна деталь не выйдет из строя
p_none = (1-p_third_detail) * (1-p_second_detail) * (1-p_first_detail) 
p_one = 1 - p_none
p_one_to_two = p_one + p_two

print('Задача 5')
print(f'Вероятность, что все детали выйдут из строя  = {round(p_all,5)}')
print(f'Вероятность, что только две детали детали выйдут из строя  = {round(p_two,5)}')
print(f'Вероятность, что хотя бы одна деталь выйдет из строя  = {round(p_one,5)}')
print(f'Вероятность, что от одной до двух деталей выйдет из строя  = {round(p_one_to_two,5)}')
```
