# Задача 2

Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).


```python
import numpy as np
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
alpha = 1e-6
B1 = 0.1
n = len(zp)
def mse_(B1, x, y, n):
    return np.sum((B1*x-y)**2/n)
for i in range(1501):
    B1 -= alpha * 2/n * np.sum((B1*zp-ks)*zp)
    
    if i % 300 == 0:
        print('Iteration = {i}, B1 = {B1}, mse = {mse}'.format(i = i+1, B1 = B1, mse = mse_(B1,zp,ks,n)))
```

**На 1201 итерации коэффициент B1 достигает значения 5.889820420132673 и дальше не меняется, полученное значение примерно равно, коэффициенту без интерцепта, вычисленному в предыдущей задаче.**

# Задача 3

Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

```python
import numpy as np
from sklearn.linear_model import LinearRegression
model = LinearRegression()
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
alpha = 1e-6
B1 = 0.1
B0 = np.mean(ks) - B1 * np.mean(zp)
n = len(zp)
def mse_(B0,B1, x, y, n):
    return np.sum((B0+B1*x-y)**2/n)
for i in range(5501):    
    B1 -= alpha * 2/n * np.sum((B0+B1*zp-ks)*zp)
    B0 = np.mean(ks) - B1 * np.mean(zp)
    if i % 500 == 0:
        print(f'Iteration = {i+1}, B0 = {B0}, B1 = {B1}, mse = {mse_(B0,B1,zp,ks,n)}')

zp = zp.reshape(-1,1)
regres = model.fit(zp,ks)
print(regres.intercept_)
print(regres.coef_)
```

**Произведен расчет коэффициентов математически и с помощью функции. Коэффициенты, расчитываемые математически, приблизительно равны коэффициентам, полученным с помощью функции и вычисленным в задаче 1.**
