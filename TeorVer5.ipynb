#–ó–∞–¥–∞—á–∞ 1#

–ö–æ–≥–¥–∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞, –∞ –∫–æ–≥–¥–∞ Z ‚Äì–∫—Ä–∏—Ç–µ—Ä–∏–π?

**–ö—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –µ—Å–ª–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞ —Å–∏–≥–º–∞ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏. –ï—Å–ª–∏ —Å–∏–≥–º–∞ –∏–∑–≤–µ—Å—Ç–Ω–∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è Z ‚Äì–∫—Ä–∏—Ç–µ—Ä–∏–π.**

```python
n=10
masses = [2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34]
m_3 = 2.5
average = sum(masses)/len(masses)
print(average)
sigma_n = 0
for i in masses:
  sigma_n += (i-average)*(i-average)
sigma_n = pow(sigma_n/(n-1),0.5)
print(sigma_n)
t_n = (average-m_3)/(sigma_n/pow(n,0.5))
print(t_n)
```

–ò—Å–ø–æ–ª—å–∑—É—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π —Å Œ±=0,05, –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —ç—Ç—É –≥–∏–ø–æ—Ç–µ–∑—É, –µ—Å–ª–∏ –≤ –≤—ã–±–æ—Ä–∫–µ –∏–∑ n=100 —à–∞—Ä–∏–∫–æ–≤ —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä –æ–∫–∞–∑–∞–ª—Å—è —Ä–∞–≤–Ω—ã–º 17.5 –º–º, –∞ –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏–∑–≤–µ—Å—Ç–Ω–∞ –∏ —Ä–∞–≤–Ω–∞ 4 –∫–≤. –º–º.

#–ó–∞–¥–∞—á–∞ 3#

–ü—Ä–æ–≤–µ–¥–∏—Ç–µ —Ç–µ—Å—Ç –≥–∏–ø–æ—Ç–µ–∑—ã. –ü—Ä–æ–¥–∞–≤–µ—Ü —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å –ø–∞—á–∫–∏ –ø–µ—á–µ–Ω—å—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 200 –≥.

–ò–∑ –ø–∞—Ä—Ç–∏–∏ –∏–∑–≤–ª–µ—á–µ–Ω–∞ –≤—ã–±–æ—Ä–∫–∞ –∏–∑ 10 –ø–∞—á–µ–∫. 

–í–µ—Å –∫–∞–∂–¥–æ–π –ø–∞—á–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190.

–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∏—Ö –≤–µ—Å–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ.

–í–µ—Ä–Ω–æ –ª–∏ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–∞, –µ—Å–ª–∏ —É—á–∏—Ç—ã–≤–∞—Ç—å, 

—á—Ç–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 99%? (–ü—Ä–æ–≤–µ—Å—Ç–∏ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç.)

*–°–∏–≥–º–∞ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞, —Å–ª–µ–¥—É–µ—Ç  –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞.*

*t —Ç–∞–±–ª–∏—á–Ω–æ–µ –¥–ª—è 9 —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ 0,995 - 3,25*

t_–Ω=(X ÃÖ - Œº)/(œÉ_–Ω/‚àön)

```python
n=10
masses = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
m_3 = 200
average = sum(masses)/len(masses)
sigma_n = 0
for i in masses:
  sigma_n += (i-average)*(i-average)
sigma_n = pow(sigma_n/(n-1),0.5)
t_n = (average-m_3)/(sigma_n/pow(n,0.5))
print(t_n)
```

t  —Ç–∞–±–ª–∏—á–Ω–æ–µ –¥–ª—è 9 —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ 0,995 (–¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–µ) = ¬±3,25

t –Ω–∞–±–ª—é–¥–∞–µ–º–æ–µ = -1,0651

**–ü–æ—Å–∫–æ–ª—å–∫—É t –Ω–∞–±–ª—é–¥–∞–µ–º–æ–µ –ª–µ–∂–∏—Ç –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ (-3,23;3,25), —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 0,995 –≤–µ—Ä–Ω–∞ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞. –°—Ä–µ–¥–Ω–∏–π –≤–µ—Å –ø–∞—á–∫–∏ –ø–µ—á–µ–Ω—å—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 200 –≥.**


#–ó–∞–¥–∞—á–∞ 4#

–ï—Å—Ç—å –ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π?

–†–æ—Å—Ç –º–∞—Ç–µ—Ä–µ–π 172, 177, 158, 170, 178,175, 164, 160, 169, 165

–†–æ—Å—Ç –≤–∑—Ä–æ—Å–ª—ã—Ö –¥–æ—á–µ—Ä–µ–π: 173, 175, 162, 174, 175, 168, 155, 170, 160


–í —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø–æ—è—Å–Ω–µ–Ω–∏—è–º–∏ –ø—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª—è, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∏ –¥–æ–ª–∂–Ω—ã —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å—Å—è –∫–∞–∫ —Å–≤—è–∑–∞–Ω–Ω—ã–µ (–¥–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è —Ä–æ—Å—Ç–∞ –º–∞—Ç–µ—Ä–µ–π –∏ –∏–º–µ–Ω–Ω–æ –∏—Ö –¥–æ—á–µ—Ä–µ–π, –Ω–µ —Å–æ—Å–µ–¥—Å–∫–∏—Ö) –∏–∑ —Ä–∞—Å—á—ë—Ç–æ–≤ –∏—Å–∫–ª—é—á—ë–Ω –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ —Ä–æ—Å—Ç–∞ –º–∞—Ç–µ—Ä–µ–π –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ ttest_rel.

–î–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞–ª–∏—á–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ t-—Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞, –¥–ª—è —á–µ–≥–æ –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π ttest_rel –∏–∑ –ø–∞–∫–µ—Ç–∞ scipy.stats.

–ü—Ä–∏–º–µ–º Œ±=0,05.

```python

from scipy.stats import ttest_rel

mothers = [172, 177, 158, 170, 178, 175, 164, 160, 169]
daughters = [173, 175, 162, 174, 175, 168, 155, 170, 160]

t_stat, p_value = ttest_rel(mothers, daughters)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```
